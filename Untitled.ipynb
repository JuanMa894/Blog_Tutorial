```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
! ls
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.drop("Outcome", axis=1)  
y = df["Outcome"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=41)
```

```python
tree = DecisionTreeClassifier(random_state=41)
```

```python
tree.fit(X_train, y_train)
```

```python
tree.tree_.max_depth
```

#### Predictions and Evaluation

```python
y_pred_test = tree.predict(X_test)
y_pred_test
```

```python
accuracy_score(y_test, y_pred_test)
```

```python
print('Training:', tree.score(X_train, y_train))
print('Testing:', tree.score(X_test, y_test))
```

```python
f, ax = plt.subplots(figsize=(10, 10))
plot_tree(tree, ax=ax);
# plt.savefig('./big_tree.png', dpi= 600)
```

```python
tree2 = DecisionTreeClassifier(min_samples_leaf=10, max_depth=8, min_samples_split=50)
```

```python
tree2.fit(X_train, y_train)
```

```python
print('Training:', tree2.score(X_train, y_train))
print('Testing:', tree2.score(X_test, y_test))
```

```python
f, ax = plt.subplots(figsize=(10, 10))
plot_tree(tree2, ax=ax);
```
